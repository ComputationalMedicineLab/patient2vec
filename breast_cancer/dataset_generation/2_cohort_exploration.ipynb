```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
```

```python
breast_cancer_df = pd.read_csv("../data/raw/breast_cancer.csv")
breast_cancer_control_df = pd.read_csv("../data/raw/breast_cancer_controls.csv")
```

# Breast cancer patients

```python
breast_cancer_df.head()
```

```python
breast_cancer_df[['COUNT_BREAST_CANCER', 'COUNT_CODES_BEFORE', 'MONTHS_CODES_BEFORE']].describe()
```

# Control patients

```python
breast_cancer_control_df.head()
```

```python
breast_cancer_control_df[['COUNT_CODES', 'MONTHS_CODES']].describe()
```

# Code counts

```python
bins = np.linspace(20, 400, 50)

plt.hist(breast_cancer_df['COUNT_CODES_BEFORE'].values, bins, normed=True, alpha=0.5, label='breast cancer')
plt.hist(breast_cancer_control_df['COUNT_CODES'].values, bins, normed=True, alpha=0.5, label='control')
plt.legend(loc='upper right')
plt.show()
```

# Number of months

```python
bins = np.linspace(12, 330, 50)

plt.hist(breast_cancer_df['MONTHS_CODES_BEFORE'].values, bins, normed=True, alpha=0.5, label='breast cancer')
plt.hist(breast_cancer_control_df['MONTHS_CODES'].values, bins, normed=True, alpha=0.5, label='control')
plt.legend(loc='upper right')
plt.show()
```

# Diabetes vs. control

```python
matplotlib.rcParams['figure.figsize'] = (15.0, 15.0)

plt.scatter(breast_cancer_control_df['COUNT_CODES'].values, breast_cancer_control_df['MONTHS_CODES'].values, alpha=0.5, label='control')
plt.scatter(breast_cancer_df['COUNT_CODES_BEFORE'].values, breast_cancer_df['MONTHS_CODES_BEFORE'].values, c='red', alpha=0.5, label='breast cancer')
plt.legend(loc='upper right')
plt.xlim((10,6000))
plt.xlabel('Code count')
plt.ylim((12,430))
plt.ylabel('Months')
plt.show()
```
