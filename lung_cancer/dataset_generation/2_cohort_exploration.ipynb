```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
```

```python
lung_cancer_df = pd.read_csv("../data/raw/lung_biopsy_cancer.csv")
lung_cancer_control_df = pd.read_csv("../data/raw/lung_biopsy_cancer_control.csv")
```

# Lung cancer patients

```python
lung_cancer_df[['CODES_BEFORE', 'MONTHS_CODES_BEFORE']].describe()
```

# Control patients

```python
lung_cancer_control_df.head()
```

```python
lung_cancer_control_df[['CODES_BEFORE', 'MONTHS_CODES_BEFORE']].describe()
```

# Code counts

```python
bins = np.linspace(20, 400, 50)

plt.hist(lung_cancer_df['CODES_BEFORE'].values, bins, normed=True, alpha=0.5, label='lung cancer')
plt.hist(lung_cancer_control_df['CODES_BEFORE'].values, bins, normed=True, alpha=0.5, label='control')
plt.legend(loc='upper right')
plt.show()
```

# Number of months

```python
bins = np.linspace(12, 330, 50)

plt.hist(lung_cancer_df['MONTHS_CODES_BEFORE'].values, bins, normed=True, alpha=0.5, label='lung cancer')
plt.hist(lung_cancer_control_df['MONTHS_CODES_BEFORE'].values, bins, normed=True, alpha=0.5, label='control')
plt.legend(loc='upper right')
plt.show()
```

# Diabetes vs. control

```python
matplotlib.rcParams['figure.figsize'] = (15.0, 15.0)

plt.scatter(lung_cancer_control_df['CODES_BEFORE'].values, lung_cancer_control_df['MONTHS_CODES_BEFORE'].values, alpha=0.5, label='control')
plt.scatter(lung_cancer_df['CODES_BEFORE'].values, lung_cancer_df['MONTHS_CODES_BEFORE'].values, c='red', alpha=0.5, label='lung cancer')
plt.legend(loc='upper right')
plt.xlim((10,6000))
plt.xlabel('Code count')
plt.ylim((12,430))
plt.ylabel('Months')
plt.show()
```
