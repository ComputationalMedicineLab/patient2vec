```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
```

```python
diabetes_patients_df = pd.read_csv("../data/raw/diabetes_drug.csv")
diabetes_patients_control_df = pd.read_csv("../data/raw/diabetes_drug_control.csv")
```

# Diabetes patients

```python
diabetes_patients_df.head()
```

```python
diabetes_patients_df[['COUNT_DIABETES_DRUG', 'COUNT_CODES_BEFORE_DRUG', 'MONTHS_CODES_BEFORE_DRUG']].describe()
```

# Control patients

```python
diabetes_patients_control_df.head()
```

```python
diabetes_patients_control_df[['COUNT_CODES', 'MONTHS_CODES']].describe()
```

# Code counts

```python
bins = np.linspace(20, 400, 50)

plt.hist(diabetes_patients_df['COUNT_CODES_BEFORE_DRUG'].values, bins, normed=True, alpha=0.5, label='diabetes')
plt.hist(diabetes_patients_control_df['COUNT_CODES'].values, bins, normed=True, alpha=0.5, label='control')
plt.legend(loc='upper right')
plt.show()
```

# Number of months

```python
bins = np.linspace(12, 330, 50)

plt.hist(diabetes_patients_df['MONTHS_CODES_BEFORE_DRUG'].values, bins, normed=True, alpha=0.5, label='diabetes')
plt.hist(diabetes_patients_control_df['MONTHS_CODES'].values, bins, normed=True, alpha=0.5, label='control')
plt.legend(loc='upper right')
plt.show()
```

# Diabetes vs. control

```python
matplotlib.rcParams['figure.figsize'] = (15.0, 15.0)

plt.scatter(diabetes_patients_control_df['COUNT_CODES'].values, diabetes_patients_control_df['MONTHS_CODES'].values, alpha=0.5, label='control')
plt.scatter(diabetes_patients_df['COUNT_CODES_BEFORE_DRUG'].values, diabetes_patients_df['MONTHS_CODES_BEFORE_DRUG'].values, c='red', alpha=0.5, label='diabetes')
plt.legend(loc='upper right')
plt.xlim((10,6000))
plt.xlabel('Code count')
plt.ylim((12,430))
plt.ylabel('Months')
plt.show()
```
